```r
meta <- data.frame (sample = c('sample1', 'sample2', 'sample3', 'sample4', 'sample5', 'sample6', 'sample7', 'sample8', 'sample9', 'sample10', 'sample11', 'sample12'), 
                sex = c('M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'),
                 stage = c('I', 'II', 'II', 'I', 'II', 'II', 'I', 'II', 'II', 'I', 'II', 'II'), 
                 treatment = rep(c('A','B','P'), each=4),
                   myc = c(2343, 457, 4593, 9035, 3450, 3524, 958, 1053, 8674, 3424, 463, 5105))
print(meta)
```

```r
#removing the numbers to the left of sample
rownames(meta) <- meta$sample
meta$sample <- NULL
print(meta)
```

```r
treatment_sex <- meta[, c('treatment', 'sex')]
print(treatment_sex)
#blank space before the comma is to select all rows we're asking for
```

```r
# return treatment values for 5,7,9, and 10
samples_selected <- c(5, 7, 9, 10)
treatments_data <- meta[samples_selected, "treatment"]
print(treatments_data)
```

```r
# use subset() to return all data for samples receiving treatment P
treatment_P_data <- subset(meta, treatment == 'P')
print(treatment_P_data)
```

```r
# print myc data over 5000 with stage and treatment
myc_over_5000 <- meta [meta$myc > 5000, c('stage', 'treatment')]
print(myc_over_5000)
```

```r
meta <- meta[, !(colnames(meta) %in% 'treatment')]
print (meta)

# remove treatment column, colnames() to get column names 
# in to check each column
# ! negates condition to remove treatment column
```

```r
# remove samples 7, 8 and 9 from the dataset

meta <- meta [ -c (7,8,9), ]
print (meta)
```

```r
# keep only samples 1-6

meta <- meta[1:6, ]
print(meta)

# , indicates all columns are kept and 1:6 calls rows with 1 through 6 
```

```r
# add a column called pre_treatment to the beginning of the dataframe
# values: T, F, F, F, T, T, F, T, F, F, T, T

pre_treatment_values <- c('T', 'F', 'F', 'F', 'T', 'T', 'F', 'T', 'F', 'F', 'T', 'T')
meta <- cbind(pre_treatment = pre_treatment_values, meta)
print(meta)
```
